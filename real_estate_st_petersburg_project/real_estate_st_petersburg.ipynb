#### locality_name

```python
# Узнаем количество пропущенных значений
data['locality_name'].isna().sum()
```

```python
data = data.dropna(subset=['locality_name'])
data['locality_name'].isna().sum()
```

```python
# Избавимся от наименования типа 

excess_words = ['деревня', 'поселок', 'посёлок', 'садовое', 'товарищество', 'коттеджный', 'городского', 'типа', 'село', 'некоммерческое', 'городской', 'станции']

for element in excess_words:
    data['locality_name'] = data['locality_name'].str.replace(element, '')

 
    
```

#### ceiling_height

Пропущенные значения в столбце 'ceiling height' заменим медианным значением по населенному пункту:

```python
print('Количество пропущенных значений "до":', data['ceiling_height'].isna().sum())    
for year in data['locality_name'].unique():
    median = data.loc[data['locality_name'] == year, 'ceiling_height'].median()
    data.loc[(data['ceiling_height'].isna()) & (data['locality_name'] == year), 'ceiling_height'] = median

print('Количество пропущенных значений "после":', data['ceiling_height'].isna().sum())    
```

Судя по всему, для некоторых неселенных пунктов вообще не указана высота потолков, заменим из медианным значением по датасету:

```python
data.loc[data['ceiling_height'].isna(), 'ceiling_height'] = data['ceiling_height'].median()
print('Количество пропущенных значений "после":', data['ceiling_height'].isna().sum())    
```

#### floors_total

Значение количества этажей также заменим медианой по населенному пункту:

```python
print('Количество пропущенных значений "до":', data['floors_total'].isna().sum())    
for year in data['locality_name'].unique():
    median = data.loc[data['locality_name'] == year, 'floors_total'].median()
    data.loc[(data['floors_total'].isna()) & (data['locality_name'] == year), 'floors_total'] = median

print('Количество пропущенных значений "после":', data['floors_total'].isna().sum())    
```

#### living_area

Пропущенные значения в столбце 'living area' заменим медианными по количеству комнат:

```python
print('Количество пропущенных значений "до":', data['living_area'].isna().sum())    
for year in data['rooms'].unique():
    median = data.loc[data['rooms'] == year, 'living_area'].median()
    data.loc[(data['living_area'].isna()) & (data['rooms'] == year), 'living_area'] = median

print('Количество пропущенных значений "после":', data['living_area'].isna().sum())    
```

#### is_apartment

В стобце с информацией о том, является ли квартира апартаменатами много пропущенных значений. Скорее всего пропущенные значения указывают на то, что квартира апартаментами не является:

```python
data['is_apartment'].isna().sum()
```

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

```python
data['is_apartment'].isna().sum()
```

#### kitchen_area

Значения в столбце 'kitchen area' заменим на медианные по количеству комнат:

```python
print('Количество пропущенных значений "до":', data['kitchen_area'].isna().sum())    
for year in data['rooms'].unique():
    median = data.loc[data['rooms'] == year, 'kitchen_area'].median()
    data.loc[(data['kitchen_area'].isna()) & (data['rooms'] == year), 'kitchen_area'] = median

print('Количество пропущенных значений "после":', data['kitchen_area'].isna().sum())
```

```python
data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = data['kitchen_area'].median()
print('Количество пропущенных значений "после":', data['kitchen_area'].isna().sum())    
```

#### airports_nearest

Расстояние до ближайщего аэропорта заменим медианными значениями по населенному пункту:

```python
print('Количество пропущенных значений "до":', data['airports_nearest'].isna().sum())    
for year in data['locality_name'].unique():
    median = data.loc[data['locality_name'] == year, 'airports_nearest'].median()
    data.loc[(data['airports_nearest'].isna()) & (data['locality_name'] == year), 'airports_nearest'] = median

print('Количество пропущенных значений "после":', data['airports_nearest'].isna().sum())
```

#### cityCenters_nearest

Расстояние до центра города также заменим медианным значением по населенному пункту:

```python
print('Количество пропущенных значений "до":', data['cityCenters_nearest'].isna().sum())    
for year in data['locality_name'].unique():
    median = data.loc[data['locality_name'] == year, 'cityCenters_nearest'].median()
    data.loc[(data['cityCenters_nearest'].isna()) & (data['locality_name'] == year), 'cityCenters_nearest'] = median

print('Количество пропущенных значений "после":', data['cityCenters_nearest'].isna().sum())
```

#### parks_around3000 and parks_nearest

```python
data[data['parks_around3000'].isna()].head()
```

Как можно заменить,пропущенные значения совпадают в стобцах 'ponds_around3000' и 'ponds_nearest'. Можно предположить, что в округе просто нет прудов. Также поступим с парками.

```python
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
```

### Изменение типа данных:

```python
data.info()
```

```python
# Приведем значения к целочисленным
columns = ['last_price', 'total_area', 'floors_total', 'living_area', 'kitchen_area', 'balcony']
for i in columns:
    data[i] = data[i].astype(int)
```

```python
#изменим столбец с датой размещения объявления на datetime:
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

```python
#заменим пропуски в этом столбце, вычтя из самой поздней даты размещения обяъявления дату, соответствующую строчкам с провещнными значения
data.loc[data['days_exposition'].isna(), 'days_exposition'] = (data['first_day_exposition'].max() - data.loc[data['days_exposition'].isna(), 'first_day_exposition']).dt.days


```

## Расчёты и добавление результатов в таблицу

### Цена квадратного метра:

```python
data['price_per_metr'] = data['last_price'] / data['total_area']
```

### Год, месяц, день недели:

```python
data['weekday_of_publ'] = data['first_day_exposition'].dt.weekday
data['month_of_publ'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['year_of_publ'] = pd.DatetimeIndex(data['first_day_exposition']).year
```

### Этаж квартиры:

```python
# Функция для присваивания типа этажа
def floor_category(row):
    if row['floor'] == 1: return 'первый'
    if row['floor'] == row['floors_total']: return 'последний'
    return 'другой'

data['floor_category'] = data.apply(floor_category, axis = 1)


data[['floors_total', 'floor_category']].head()
```

### Соотношение жилой и общей площади, а также отношение площади кухни к общей:

```python
data['living_to_total'] = data['living_area'] / data['total_area']
data['kitchen_to_total'] = data['kitchen_area'] / data['total_area']
```
